In the *naive* approach to object segmentation, we used the sliding window technique to generate a region of interest (ROIs) for an image. But this approach was not practical for our case of plane detection, it relayed only on one dimension of the window, and most of all - it was very slow!

Before we build our R-CNN model (in the next lesson), we need to learn a better, smarter way of finding ROIs! That's where the Selective Search algorithm comes. 

The selective Search algorithm handles the issue with different scales that we had. Still, it looks at the texture of certain parts of an image and, based on that group of items, produces regions and ultimately proposes ROIs. Because of the grouping part, this algorithm works very fast!
